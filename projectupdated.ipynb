```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel(r"C:\Users\dell\Downloads/Online_Retail.xlsx")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

## REMOVING CANCELLED DATA


```python
df=df[~df["InvoiceNo"].astype(str).str.contains("C", na=False)]
df
```

# REMOVING EMPTY DATA IN CUSTOMER ID

```python
df=df.dropna(subset=["CustomerID"])
df
```

## SALES

```python
df["Sales"]= df["Quantity"] * df["UnitPrice"]
df
#here we multiplied quantity and unit price to get sales/revenue
```

```python
df["InvoiceDate"]= pd.to_datetime(df["InvoiceDate"])
# changed the invoicedate in the original data to datetime to enable editing
```

```python
df["Month"]=df["InvoiceDate"].dt.month_name()
# extraction of month from the data
```

```python
df["Day"]=df["InvoiceDate"].dt.day_name()
#extraction of day in a week from the data
```

```python
df["Year"]=df["InvoiceDate"].dt.year
# extraction of the year
```

```python
df["Week"]=df["InvoiceDate"].dt.isocalendar().week
# extraction of the week number
```

```python
df
```

## TOTAL SALES MONTHLY

```python
Monthly_sales= df.groupby("Month")["Sales"].sum().reset_index()
Monthly_sales.sort_values(by="Sales",ascending=False)
# we first of all grouped the data into months and total number of sales that occured in each month
# arranging the months from the highest sales down to the lowest sales
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x="Month",y="Sales",data=Monthly_sales)
plt.grid()
plt.title("TOTAL MONTHLY SALES")
plt.savefig("TOTAL MONTHLY SALES.png")
```

### TOTAL SALES WEEKLY

```python
Weekly_sales= df.groupby("Week")["Sales"].sum().reset_index()
Weekly_sales
# grouping the data set into each week number and total sales that occured in each week
```

```python
Weekly_sales2=Weekly_sales.sort_values(by="Sales",ascending=False)
Weekly_sales2.head(10)
# arranging the weekly sales from the highest to the least then extracting the top 10
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x="Week",y="Sales",data=Weekly_sales2.head(10))
plt.grid()
plt.title("TOTAL WEEKLY SALES")
plt.savefig("TOTAL WEEKLY SALES.png")
```

# TOP CUSTOMERS BY QUANTITY AND SALES


```python
Top_customers_by_quantity= df.groupby("CustomerID").agg({"Quantity":"sum",
                                                        "Sales":"sum"}).reset_index()
Tcq_10=Top_customers_by_quantity.sort_values(by="Sales",ascending=False)
```

```python
Tcq_10
```

# FINDING OUT THE FREQUENT AND RARE BUYERS


```python
order=df.groupby("CustomerID")["InvoiceNo"].nunique().reset_index()
bins =[0,5,300]
order["Frequency"]=pd.cut(order["InvoiceNo"],bins=bins,labels=["rarely","loyal"])
```

```python
loyal_customers=order.sort_values(by="InvoiceNo",ascending=False)
loyal_customers
```

```python
top_fre=loyal_customers.merge(Tcq_10)
```

```python
top_frequently=top_fre.sort_values(by="Sales",ascending=False).head(10)
top_frequently
```

# SUGGESTION
* ## Frequent buyers should be rewarded discounts once in a while, should be given early access to new products
  


# ANALYSES TO KNOW HOW LONG EACH CUSTOMER MADE THEIR LAST PURCHASE (RECENCY)


```python
day_of_last_purchase=df["InvoiceDate"].max() + pd.Timedelta(days=1)
day_of_last_purchase
```

```python
customers__last_purchase=df.groupby("CustomerID")["InvoiceDate"].max().reset_index()
customers__last_purchase
```

```python
customers__last_purchase["Recency"]=(day_of_last_purchase-customers__last_purchase["InvoiceDate"]).dt.days
```

```python
customers__last_purchase
```

```python
top_frequently.merge(customers__last_purchase, on="CustomerID")
```

# SUGGESTION
## Customers who havent made purchase in a long time should be reached out to inorder to inform them of available products and promo.

# AVERAGE BASKET SIZE (PRODUCTS PER TRANSATION)


```python
Average_basket_size= df.groupby("InvoiceNo")["Quantity"].sum().mean()
Average_basket_size
```

```python
Item_bought_per_transaction=df.groupby("InvoiceNo")["Quantity"].sum().reset_index()
Item_bought_per_transaction
```

```python
Item_bought_per_transaction_10=Item_bought_per_transaction.sort_values(by="Quantity",ascending=False).head(10)
Item_bought_per_transaction_10
```

```python
High_quantity_invoice=Item_bought_per_transaction[Item_bought_per_transaction ["Quantity"] >= Average_basket_size]
High_quantity_invoice.count()
```

```python
Low_quantity_invoice=Item_bought_per_transaction[Item_bought_per_transaction ["Quantity"] < Average_basket_size]
Low_quantity_invoice.count()
```

## STRATEGY TO INCREASE BASKET SIZE
* To increase basket size, promo/ combo offers should be initiated. for instance, buy a specific quantity of products, get a discount or one free.
* Products that are related to the goods purchased by customers should be displayed incase the customer is in need of that same product but isn't aware that the company sells such(recommend related items).
